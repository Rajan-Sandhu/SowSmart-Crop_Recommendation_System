```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
crop = pd.read_csv("data/Crop_recommendation.csv")
```

```python
crop.head()
```

```python
crop.isna().sum()
```

```python
crop.shape
```

```python
# sns.pairplot(crop,hue="label")
```

```python
X=crop.drop("label",axis=1)
X.head()
```

```python
Y=crop.loc[:,"label"]
Y
```

```python
# Correlation heatmap
numeric_df = crop.select_dtypes(include=[np.number])
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='rocket')
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
model3=KNeighborsClassifier()
model3.fit(x_train,y_train)
```

```python
y_pred=model3.predict(x_test)
y_pred
```

```python
compare=pd.DataFrame({"Actual":y_test,"Predicted":y_pred})
compare.head()
```

```python
# Accuracy
print("Accuracy: ",accuracy_score(y_test,y_pred))
```

```python
# Confusion Matrix
print("Confusion Matrix: \n",confusion_matrix(y_test,y_pred))
```

```python
# class_report
print("Classification_report: \n",classification_report(y_test,y_pred))
```

```python
import joblib
joblib.dump(model3,"models/crop_knn_model.pkl")
```
